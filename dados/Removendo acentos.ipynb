```python
import pandas as pd
from unidecode import unidecode

# ler o arquivo CSV com os dados separados por ponto e vírgula
inspecao = pd.read_csv("inspecao.csv", delimiter=";")

# remover acentos do nome das colunas
inspecao.columns = [unidecode(col) for col in inspecao.columns]

# remover acentos do conteúdo das colunas
for col in inspecao.columns:
    if inspecao[col].dtype == "object":
        inspecao[col] = inspecao[col].apply(lambda x: unidecode(x))



```

```python
inspecao
```

```python
# substituir "Inspeção" por "Inspecao" em todas as células da coluna de índice 5
inspecao.iloc[:, 5] = inspecao.iloc[:, 5].apply(lambda x: x.replace("Inspeo", "Inspecao"))


```

```python
#salvar as alterações em um novo arquivo

inspecao.to_csv("inspecao_sem_acentos.csv", sep=";", index=False)
```
